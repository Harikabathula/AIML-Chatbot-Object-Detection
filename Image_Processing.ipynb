```python
import cv2
from google.colab.patches import cv2_imshow

# Read the color image
color_image = cv2.imread('/content/Blue_Image.jpg')

# Convert the color image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray_image)

# Convert the grayscale image back to color
color_image_restored = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)

# Display the restored color image
cv2_imshow(color_image_restored)

# Save the restored color image
cv2.imwrite('/content/restored_color_image.jpg', color_image_restored)
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image
image = cv2.imread('/content/Blue_Image.jpg')
cv2_imshow(image)

# Split the image into color channels
blue, green, red = cv2.split(image)

# Set all blue pixels to red
red[blue > 0] = 255

# Merge the modified channels back into the image
modified_image = cv2.merge((blue, green, red))

# Display the modified image
cv2_imshow(modified_image)

# Save the modified image
cv2.imwrite('/content/modified_image.jpg', modified_image)
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the color image
color_image = cv2.imread('/content/Blue_Image.jpg')

# Convert the color image to negative
negative_image = 255 - color_image

# Display the negative image
cv2_imshow(negative_image)

# Save the negative image
cv2.imwrite('/content/negative_image.jpg', negative_image)
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Load the pre-trained face detection model from OpenCV
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the input image
image = cv2.imread('/content/Blue_Image.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform face detection
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around the detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with detected faces
cv2_imshow(image)

# Save the image with detected faces
cv2.imwrite('/content/detected_faces.jpg', image)
```
